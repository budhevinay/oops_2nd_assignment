```python
#19.Discuss the concept of "is-a" and "has-a" relationships in inheritance, and provide examples of each.
'''"Inheritance" is a fundamental concept in object-oriented programming (OOP), and it's often used to 
model relationships between classes. Two common types of relationships in inheritance are "is-a" and "has-a"
relationships, and they represent different ways in which classes can be related to one another.

"Is-a" Relationship (Inheritance):

An "is-a" relationship represents a subclass inheriting from a superclass, indicating that the subclass is a
specialized version of the superclass. In other words, the subclass shares the characteristics and behaviors 
of the superclass while adding its own specific features.'''
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

my_dog = Dog()
my_cat = Cat()

print(my_dog.speak())  
print(my_cat.speak())  
''''
"Has-a" Relationship (Composition):

A "has-a" relationship represents a class that contains an instance of another class as one of its attributes.
This relationship is based on composition, where one class has another class as a component to fulfill its 
functionality.'''
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()  

    def drive(self):
        print("Car is driving")

my_car = Car()
my_car.drive()
my_car.engine.start()
'''In this example, the Car class has a "has-a" relationship with the Engine class. The Car class contains 
an instance of the Engine class as an attribute, and it can access and use the Engine object to start the 
engine.'''
```

```python
#2. Describe the key principles of encapsulation, including access control and data hiding.
'''Encapsulation is one of the core principles of object-oriented programming (OOP) and involves two key 
concepts: access control and data hiding. These principles help create a well-structured and maintainable 
codebase. 
1.Access Control:
Access control defines how and where an object's attributes and methods can be accessed. It determines who 
can use and modify an object's internal state

There are typically three levels of access control:

    1.Public: Members (attributes and methods) declared as public are accessible from anywhere in the program. 
    2.Protected: Members declared as protected are accessible within the class,its subclasses (derived classes), 
      and sometimes within the same package or module, but not from external code.
    3.Private: Members declared as private are intended to be accessed only within the class that defines them. 
      They are hidden from external code
2.Data Hiding:
Data hiding is the practice of restricting direct access to an object's internal state (attributes). The 
primary goal of data hiding is to protect the integrity of the object's data and to ensure that changes to
the data occur in a controlled and predictable manner. This is achieved through encapsulation by making the 
object's attributes private or using accessors (getter methods) and mutators (setter methods) to interact
with the attributes.

Data hiding offers several benefits:
    1.Encapsulation of Behavior: Besides data, encapsulation also includes encapsulating the behavior (methods)
    associated with an object. This helps in maintaining a clear and structured way to manipulate the object's 
    data.

    2.Security and Validation: With data hiding, you can implement security measures and validation checks 
    within the accessor and mutator methods to ensure that the data remains consistent and valid.'''
```

```python
#3. How can you achieve encapsulation in Python classes? Provide an example.
'''Achieving encapsulation in Python classes involves using access specifiers and defining appropriate methods
to control access to the class's attributes. 

In Python, there are three levels of access control:

1.Public: Members are accessible from anywhere.
2.Protected: Members should not be accessed from outside the class, but they can be accessed in subclasses.
3.Private: Members should not be accessed from outside the class. In Python, they are denoted by prefixing 
the member name with double underscores.'''
class Student:
    def __init__(self, name, age):
        self.__name = name  # Private attribute
        self.__age = age  # Private attribute

    # Accessor methods to get the values of private attributes
    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    # Mutator methods to set the values of private attributes with validation
    def set_name(self, name):
        if isinstance(name, str):
            self.__name = name
        else:
            print("Name must be a string.")

    def set_age(self, age):
        if isinstance(age, int) and 0 <= age <= 150:
            self.__age = age
        else:
            print("Age must be an integer between 0 and 150.")

# Create a Student object
student = Student("Alice", 20)

# Accessing private attributes through accessor methods
print("Student Name:", student.get_name())
print("Student Age:", student.get_age())

# Modifying private attributes through mutator methods
student.set_name("Bob")
student.set_age(25)

# Attempting to set invalid values
student.set_name(123)  # Invalid name
student.set_age(-5)    # Invalid age
```

```python
#12.Explain the concept of property decorators in Python and how they relate to encapsulation.
'''Property decorators in Python are a powerful feature that allows you to define getter and setter methods 
for class attributes in a more concise and intuitive way. They are used to encapsulate the access to class 
attributes, making it easier to implement controlled access to an object's data. 
1.@property:

The @property decorator is used to define a method as a "getter" for a class attribute.
It allows you to access the attribute like a regular attribute without using parentheses as if it were a
method.'''
class MyClass:
    def __init__(self):
        self.__my_variable = 42  # Private attribute

    @property
    def my_variable(self):
        return self.__my_variable

obj = MyClass()
value = obj.my_variable  # Access the attribute like a property
'''
2.@<attribute>.setter:

The @<attribute>.setter decorator is used to define a method as a "setter" for a class attribute.
It allows you to control how the attribute is modified by defining custom logic and validation.
It's executed when you assign a new value to the attribute.'''
class MyClass:
    def __init__(self):
        self.__my_variable = 23  # Private attribute

    @property
    def my_variable(self):
        return self.__my_variable

    @my_variable.setter
    def my_variable(self, new_value):
        if new_value >= 0:
            self.__my_variable = new_value

obj = MyClass()
obj.my_variable = 50  
```

```python
#2. Describe the difference between compile-time polymorphism and runtime polymorphism in Python.
'''There are two main types of polymorphism in Python:

Compile-Time Polymorphism (Static Binding): This is also known as method overloading. In this type of 
polymorphism, multiple methods in the same class have the same name but different parameter lists. 
The appropriate method to be called is determined at compile-time based on the method signature.

Run-Time Polymorphism (Dynamic Binding): This is also known as method overriding. In this type of 
polymorphism, a subclass can provide a specific implementation of a method that is already defined in its
superclass. The appropriate method to be called is determined at run-time based on the actual object's class.'''
```

```python
'''3. Create a Python class hierarchy for shapes (e.g., circle, square, triangle) and demonstrate polymorphism 
through a common method, such as `calculate_area()`.'''
import math
class shape:
    def area(self):
        pass
class circle(shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return math.pi*self.radius**2
    
class square(shape):
    def __init__(self,length):
        self.length=length
    def area(self):
        return self.length*self.length
class triangle(shape):
    def __init__(self,base,height):
        self.base=base
        self.height=height
    def area(self):
        return 0.5*self.base*self.height
sqr=square(4)
sqr.area()
print(sqr.area())
cir=circle(5)
cir.area()
print(cir.area())
```

```python
#4. Explain the concept of method overriding in polymorphism. Provide an example.
'''Run-Time Polymorphism (Dynamic Binding): This is also known as method overriding. In this type of 
polymorphism, a subclass can provide a specific implementation of a method that is already defined in its
superclass. The appropriate method to be called is determined at run-time based on the actual object's class.'''
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof"

class Cat(Animal):
    def speak(self):
        return "Meow"

def animal_sound(animal):
    return animal.speak()

dog = Dog()
cat = Cat()

print(animal_sound(dog))  
print(animal_sound(cat))  
```

```python
'''4. Explain the concept of abstract classes in Python and how they are defined using the `abc` module. 
Provide an example.'''

'''Abstract classes in Python are classes that cannot be instantiated on their own and are meant to be 
subclassed by other classes. They often serve as a blueprint for other classes, defining a common interface 
or a set of methods that must be implemented by their subclasses. Abstract classes may also contain abstract
methods, which are methods without a defined implementation in the abstract class. Subclasses are required to
provide concrete implementations for these abstract methods.

In Python, abstract classes can be defined using the abc module (Abstract Base Classes).'''
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

    @abstractmethod
    def display_info(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14 * self.radius**2

    def display_info(self):
        print(f"Circle with radius {self.radius}")

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def display_info(self):
        print(f"Rectangle with length {self.length} and width {self.width}")


circle = Circle(radius=5)
rectangle = Rectangle(length=4, width=6)


print("Circle Area:", circle.calculate_area())
circle.display_info()

print("Rectangle Area:", rectangle.calculate_area())
rectangle.display_info()
```

```python
#5. How do abstract classes differ from regular classes in Python? Discuss their use cases.
'''Abstract classes in Python differ from regular classes in that abstract classes cannot be instantiated on
their own, and they may contain abstract methods that lack an implementation in the abstract class itself. 
Regular classes, on the other hand, can be instantiated, and they provide concrete implementations for all 
their methods.

Here are some key differences between abstract classes and regular classes in Python:
1.Instantiation:

Abstract classes cannot be instantiated directly. They are meant to be subclassed, and their abstract methods
must be implemented by the subclasses.
Regular classes can be instantiated directly, and objects can be created from them.

2.Abstract Methods:

Abstract classes can have abstract methods, which are methods without a defined implementation in the abstract
class. Subclasses are required to provide concrete implementations for these abstract methods.
Regular classes do not have abstract methods. All methods in a regular class have concrete implementations.

3.Purpose and Use Cases:

Abstract classes are often used to define a common interface for a group of related classes. They provide a 
blueprint for subclasses to follow, ensuring that specific methods are implemented in each subclass.
Regular classes are used when you want to create objects with concrete implementations for all methods.
They represent specific entities or concepts in your program.

4.Enforcement of Implementation:

Abstract classes enforce the implementation of certain methods by making them abstract. If a subclass does 
not provide concrete implementations for all abstract methods, attempting to instantiate the subclass will 
result in a TypeError.
Regular classes do not enforce the implementation of any specific methods. It is possible to create instances
of a regular class even if some methods are not implemented.'''
```

```python
'''6.Create a Python class for a bank account and demonstrate abstraction by hiding the account balance and 
providing methods to deposit and withdraw funds.'''
from abc import ABC, abstractmethod

class BankAccount(ABC):
    def __init__(self, account_holder, account_number):
        self.account_holder = account_holder
        self.account_number = account_number
        self._balance = 0  

    @property
    def balance(self):
        return self._balance

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

class SavingsAccount(BankAccount):
    def __init__(self, account_holder, account_number, interest_rate=0.02):
        super().__init__(account_holder, account_number)
        self.interest_rate = interest_rate

    def deposit(self, amount):
        self._balance += amount
        print(f"Deposited ${amount}. New balance: ${self._balance}")

    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self._balance}")
        else:
            print("Insufficient funds!")

    def add_interest(self):
        interest = self._balance * self.interest_rate
        self._balance += interest
        print(f"Interest added: ${interest}. New balance: ${self._balance}")


savings_account = SavingsAccount(account_holder="John Doe", account_number="123456")
print(f"Initial balance: {savings_account.balance}")

savings_account.deposit(1000)
savings_account.withdraw(500)
savings_account.add_interest()
```

```python
'''16. Create a Python class for a computer system, demonstrating abstraction by defining common methods
(e.g., `power_on()`, `shutdown()`) in an abstract base class.'''
from abc import ABC, abstractmethod


class Computer(ABC):
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.powered_on = False

    @abstractmethod
    def power_on(self):
        pass

    @abstractmethod
    def shutdown(self):
        pass

class Laptop(Computer):
    def __init__(self, brand, model, battery_percentage):
        super().__init__(brand, model)
        self.battery_percentage = battery_percentage

    def power_on(self):
        if not self.powered_on:
            print(f"{self.brand} {self.model} laptop is booting up.")
            self.powered_on = True
        else:
            print(f"{self.brand} {self.model} laptop is already powered on.")

    def shutdown(self):
        if self.powered_on:
            print(f"{self.brand} {self.model} laptop is shutting down.")
            self.powered_on = False
        else:
            print(f"{self.brand} {self.model} laptop is already powered off.")

class Desktop(Computer):
    def __init__(self, brand, model):
        super().__init__(brand, model)

    def power_on(self):
        if not self.powered_on:
            print(f"{self.brand} {self.model} desktop is booting up.")
            self.powered_on = True
        else:
            print(f"{self.brand} {self.model} desktop is already powered on.")

    def shutdown(self):
        if self.powered_on:
            print(f"{self.brand} {self.model} desktop is shutting down.")
            self.powered_on = False
        else:
            print(f"{self.brand} {self.model} desktop is already powered off.")


laptop = Laptop(brand="Dell", model="XPS 13", battery_percentage=80)
desktop = Desktop(brand="HP", model="Pavilion")


laptop.power_on()
laptop.shutdown()

desktop.power_on()
desktop.shutdown()
```

```python
#19.Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.
from abc import ABC, abstractmethod


class Library(ABC):
    def __init__(self):
        self.books_available = {}

    @abstractmethod
    def add_book(self, book_title, quantity):
        pass

    @abstractmethod
    def borrow_book(self, book_title, user_id):
        pass


class StudentLibrary(Library):
    def add_book(self, book_title, quantity):
        if book_title in self.books_available:
            self.books_available[book_title] += quantity
        else:
            self.books_available[book_title] = quantity

    def borrow_book(self, book_title, user_id):
        if book_title in self.books_available and self.books_available[book_title] > 0:
            print(f"Student {user_id} borrowed '{book_title}'.")
            self.books_available[book_title] -= 1
        else:
            print(f"Book '{book_title}' is not available for students.")

class FacultyLibrary(Library):
    def add_book(self, book_title, quantity):
        if book_title in self.books_available:
            self.books_available[book_title] += quantity
        else:
            self.books_available[book_title] = quantity

    def borrow_book(self, book_title, user_id):
        print(f"Faculty {user_id} borrowed '{book_title}'.")
      
    
student_library = StudentLibrary()
faculty_library = FacultyLibrary()


student_library.add_book("Introduction to Python", 5)
student_library.borrow_book("Introduction to Python", user_id="S123")

faculty_library.add_book("Advanced Algorithms", 3)
faculty_library.borrow_book("Advanced Algorithms", user_id="F456")
```

```python
#15. Explain how composition enhances code maintainability and modularity in Python programs.
'''
1.Separation of Concerns:

Composition allows you to break down a complex problem into smaller, more manageable pieces. Each class is 
responsible for a specific aspect or functionality, promoting a separation of concerns.
2.Modularity:

With composition, you can design classes to be modular and reusable. Each class represents a distinct entity
or functionality, making it easier to understand, modify, and reuse in other parts of your codebase.

3.Code Reusability:

By composing classes instead of using inheritance for everything, you can reuse classes in different contexts. This promotes the creation of flexible, modular components that can be easily integrated into various parts of your system.

4.Flexibility:

Composition allows you to change the behavior of a class by substituting or adding components dynamically.
This flexibility is crucial for adapting to changing requirements without modifying existing code extensively.

5.Easier Testing:

Components created through composition can often be tested in isolation. This is because each class is
designed to do a specific task, and you can test its functionality independently of other components. This makes it easier to identify and fix issues.

6.Encapsulation:

Composition promotes encapsulation, as each class is responsible for its own behavior and data. This 
encapsulation makes it easier to reason about and maintain code, as the internal details of a class are hidden from the rest of the system.

7.Reduced Dependency:

When using composition, classes are loosely coupled, meaning they depend on abstractions rather than concrete 
implementations. This reduces the dependency between classes, making it easier to modify one class without affecting others.

8.Scalability:

As your codebase grows, composition allows you to add new features or components without having to modify 
existing code extensively. This scalability is crucial for maintaining and extending a codebase over time.'''
```

```python
#16. Create a Python class for a computer game character, using composition to represent attributes like weapons, armor, and inventory.


class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def __str__(self):
        return f"{self.name} (Damage: {self.damage})"


class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def __str__(self):
        return f"{self.name} (Defense: {self.defense})"


class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def display_inventory(self):
        print("Inventory:")
        for item in self.items:
            print(item)

    def __str__(self):
        return "Character's Inventory"


class GameCharacter:
    def __init__(self, name, health):
        self.name = name
        self.health = health
        self.weapon = None
        self.armor = None
        self.inventory = Inventory()

    def equip_weapon(self, weapon):
        self.weapon = weapon

    def equip_armor(self, armor):
        self.armor = armor

    def attack(self, target):
        if self.weapon:
            damage_dealt = self.weapon.damage
            print(f"{self.name} attacks {target.name} with {self.weapon.name} for {damage_dealt} damage!")
            target.take_damage(damage_dealt)
        else:
            print(f"{self.name} attacks {target.name} with bare hands!")

    def take_damage(self, damage):
        if self.armor:
            damage -= self.armor.defense
            if damage < 0:
                damage = 0
        self.health -= damage
        print(f"{self.name} takes {damage} damage. Health: {self.health}")

    def loot_item(self, item):
        self.inventory.add_item(item)
        print(f"{self.name} loots {item} and adds it to the inventory.")

    def display_character_info(self):
        print(f"{self.name}'s Character Info:")
        print(f"Health: {self.health}")
        if self.weapon:
            print(f"Weapon: {self.weapon}")
        if self.armor:
            print(f"Armor: {self.armor}")
        self.inventory.display_inventory()

    def __str__(self):
        return f"GameCharacter - {self.name}"

# Example usage:

# Create weapons and armor
sword = Weapon("Sword", 20)
shield = Armor("Shield", 10)

# Create game characters
player = GameCharacter("Hero", 100)
enemy = GameCharacter("Monster", 50)

# Equip weapons and armor
player.equip_weapon(sword)
player.equip_armor(shield)

# Display character info
player.display_character_info()

# Attack the enemy
player.attack(enemy)
enemy.display_character_info()

# Loot an item
potion = "Health Potion"
player.loot_item(potion)
player.display_character_info()
```

```python
#18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.

class Furniture:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"{self.name}"


class Appliance:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"{self.name}"


class Room:
    def __init__(self, name, area):
        self.name = name
        self.area = area
        self.furniture = []
        self.appliances = []

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def display_contents(self):
        print(f"{self.name} Contents:")
        print("Furniture:")
        for item in self.furniture:
            print(f"  - {item}")
        print("Appliances:")
        for item in self.appliances:
            print(f"  - {item}")

    def __str__(self):
        return f"Room - {self.name} (Area: {self.area} sq. meters)"


class House:
    def __init__(self, name):
        self.name = name
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def display_house_contents(self):
        print(f"{self.name} Contents:")
        for room in self.rooms:
            room.display_contents()

    def __str__(self):
        return f"House - {self.name}"

# Example usage:

# Create furniture and appliances
couch = Furniture("Sofa")
table = Furniture("Coffee Table")
oven = Appliance("Oven")
tv = Appliance("TV")

# Create rooms
living_room = Room("Living Room", 25)
kitchen = Room("Kitchen", 20)

# Add furniture and appliances to rooms
living_room.add_furniture(couch)
living_room.add_furniture(table)
living_room.add_appliance(tv)

kitchen.add_appliance(oven)

# Create a house and add rooms
my_house = House("Cozy Cottage")
my_house.add_room(living_room)
my_house.add_room(kitchen)

# Display the contents of the house
my_house.display_house_contents()
```

```python
#20. Create a Python class for a social media application, using composition to represent users, posts, and comments.
class Comment:
    def __init__(self, user, content):
        self.user = user
        self.content = content

    def __str__(self):
        return f"{self.user}: {self.content}"


class Post:
    def __init__(self, user, content):
        self.user = user
        self.content = content
        self.comments = []

    def add_comment(self, user, content):
        comment = Comment(user, content)
        self.comments.append(comment)

    def display_comments(self):
        print(f"Comments on post by {self.user}:")
        for comment in self.comments:
            print(comment)

    def __str__(self):
        return f"{self.user}'s Post: {self.content}"


class User:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def create_post(self, content):
        post = Post(self.username, content)
        self.posts.append(post)
        return post

    def display_posts(self):
        print(f"{self.username}'s Posts:")
        for post in self.posts:
            print(post)

    def __str__(self):
        return f"User - {self.username}"

class SocialMediaApp:
    def __init__(self, name):
        self.name = name
        self.users = []

    def add_user(self, username):
        user = User(username)
        self.users.append(user)
        return user

    def display_users(self):
        print(f"Users on {self.name}:")
        for user in self.users:
            print(user)

# Example usage:

# Create a social media app
my_social_media = SocialMediaApp("MySocialApp")

# Add users
user1 = my_social_media.add_user("Alice")
user2 = my_social_media.add_user("Bob")

# Users create posts
post1 = user1.create_post("Hello, everyone!")
post2 = user2.create_post("Just had an amazing day!")

# Users add comments to posts
post1.add_comment(user2.username, "Nice post, Alice!")
post2.add_comment(user1.username, "Glad to hear that, Bob!")

# Display users and their posts
my_social_media.display_users()

# Display posts and comments
user1.display_posts()
post1.display_comments()

user2.display_posts()
post2.display_comments()
```
